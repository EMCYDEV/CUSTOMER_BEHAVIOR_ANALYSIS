```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Create a column age group
label=['Young adult', 'Adult', 'Middle age', 'Senior']
df['Age group']= pd.qcut(df['Age'], q=4, labels=label)
```

```python
df[['Age', 'Age group']]
```

```python
# Creating purchase frequency in days
frequency_days= {
    "Fortnightly":14,
"Weekly":7,
"Quartaly": 90,
"Bi-Weekly":14,
"Annually": 365,
"Every 3 Month":90
}

df['Purchase frequency days']= df['Frequency of Purchases'].map(frequency_days)
```
